```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\saura\Downloads\Dataset\Expanded_data_with_more_features.csv")
print(df.head())
```

```python
df = df.drop("Unnamed: 0", axis = 1)
print(df.head())
```

```python
plt.figure(figsize= (6,6))
ax = sns.countplot(data = df, x = "Gender")
plt.title("Gender Distribution")
ax.bar_label(ax.containers[0])
plt.show()
```

From the above chart we have analysed that:
the number of females in the data is more than the number of males

```python
gb = df.groupby("ParentEduc").agg({"MathScore": "mean", "ReadingScore": "mean", "WritingScore" : "mean"})
print(gb)
```

```python
plt.figure(figsize= (6,6))
sns.heatmap(gb, annot = True)
plt.title("Relationship between Parent's Education and Student's Score")
plt.show()
```

From the above chart we have concluded that the education of the parents have a good impact on student's score.

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore": "mean", "ReadingScore": "mean", "WritingScore" : "mean"})
print(gb1)
```

```python
plt.figure(figsize= (6,6))
sns.heatmap(gb1, annot = True)
plt.title("Relationship between Parent's Marital Status and Student's Score")
plt.show()
```

From the above chart we have concluded that the Marital Status of the parents have a no/negligible impact on student's score.

```python
gb2 = df.groupby("IsFirstChild").agg({"MathScore": "mean", "ReadingScore": "mean", "WritingScore" : "mean"})
print(gb2)
```

From the above data we have concluded that whether it is first child or not that have a no impact on student's score.

```python
gb3 = df.groupby("PracticeSport").agg({"MathScore": "mean", "ReadingScore": "mean", "WritingScore" : "mean"})
print(gb3)
```

From the above data we have concluded that their sports practice has no effect on their score.

```python
gb4 = df.groupby("NrSiblings").agg({"MathScore": "mean", "ReadingScore": "mean", "WritingScore" : "mean"})
print(gb4)
```

From the above data we have concluded that the number of the sibbligs have a no/negligible impact on student's score.

```python
gb5 = df.groupby("WklyStudyHours").agg({"MathScore": "mean", "ReadingScore": "mean", "WritingScore" : "mean"})
print(gb5)
```

```python
From the above data we have concluded that the weekly study hours have a no/negligible impact on student's score.
```

```python
groupA = df.loc[(df[ 'EthnicGroup'] == "group A")].count()
groupB = df.loc[(df[ 'EthnicGroup'] == "group B")].count()
groupC = df.loc[(df[ 'EthnicGroup'] == "group C")].count()
groupD = df.loc[(df[ 'EthnicGroup'] == "group D")].count()
groupE = df.loc[(df[ 'EthnicGroup'] == "group E")].count()
l = ["group A", "group B", "group C","group D", 'group E']
mlist = [groupA["EthnicGroup"], groupB["EthnicGroup"], groupC["EthnicGroup"], groupD["EthnicGroup"], groupE["EthnicGroup"]]
myexplode = [0, 0, 0.1, 0, 0]
plt.pie(mlist, labels = l, autopct = "% 1.2f%%", explode = myexplode, shadow = True)
plt.title("Distribution of Ethnic Groups")
plt.show()
```
